### 2. Exibir somente os números negativos contidos no Vetor

```python
vetor = [45,-89,32,-12,33]
listNegativo = []
for indice in range(len(vetor)):
        if vetor[indice] < 0:
                listNegativo.append(vetor[indice])
                vetor[indice] = 0
print(vetor)
print(listNegativo)
```

### 3. Retornar a soma dos elementos do Vetor

```python
vetor = [45,-89,32,-12,33]
soma  = 0
for elemento in vetor:
    soma += elemento
print(soma)
```

### 4. Retornar a média dos elementos do Vetor


```python
vetor = [45,-89,32,-12,33]
soma  = 0
for elemento in vetor:
    soma += elemento
print(soma)

media = soma / len(vetor)
print(media)
```
